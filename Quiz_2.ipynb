```python
import pandas as pd
```

```python
test = pd.read_csv('test_soal_REMEDI.csv')
test
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D

model = Sequential()
model.add(Conv2D(8, (3,3), input_shape =(64,64,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(16, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.load_weights('quiz1.h5')
```

```python
image_id = test['nama_file']
test.head(10)
```

```python
from tensorflow.keras.preprocessing.image import load_img
import numpy as np

# making a list for our predictions
labels = []


for i in image_id:
    image = load_img('test_soal/'+str(i), target_size=(64, 64))
    img = np.array(image)
    img = img / 255.0
    img = img.reshape(1,64,64,3)
    label = model.predict(img)
    label_id = label[0].tolist()

    labels.append(label_id.index(max(label_id)))
```

```python
test['status'] = labels
test.head(10)
```

```python
test.to_csv('20102298.csv', columns=['status'],index=False)
```
