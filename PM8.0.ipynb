```python
import pandas as pd
import numpy as np
```

```python
bahasa_ah = pd.read_fwf('bahasa_ah.txt', header=None, delimiter=';').values[:,0]
bahasa_ah
```

```python
bahasa_uh = pd.read_fwf('bahasa_uh.txt', header=None, delimiter=';').values[:,0]
bahasa_uh
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
token_ah = Tokenizer()
token_uh = Tokenizer()

token_ah.fit_on_texts(bahasa_ah)
token_uh.fit_on_texts(bahasa_uh)
v_bahasa_ah = token_ah.texts_to_sequences(bahasa_ah)
v_bahasa_uh = token_uh.texts_to_sequences(bahasa_uh)
v_bahasa_ah
```

```python
v_bahasa_uh
```

```python
print(token_ah.texts_to_sequences(['ah ah ah uh']))
print(token_ah.sequences_to_texts([[1, 1, 1, 2, 3]]))
print(token_ah.word_index)
print(len(token_ah.word_index))
```

```python
total_kata_bahasa_ah = len(token_ah.word_index)
total_kata_bahasa_uh = len(token_uh.word_index)
```

```python
v_bahasa_ah = pad_sequences(v_bahasa_ah, maxlen = 20, padding = 'post')
v_bahasa_ah
```

```python
v_bahasa_uh = pad_sequences(v_bahasa_uh, maxlen = 20, padding = 'post')
v_bahasa_uh
```

```python
v_bahasa_ah.shape
```

```python
v_bahasa_uh.shape
```

```python
v_bahasa_ah = v_bahasa_ah.reshape(v_bahasa_ah.shape[0], v_bahasa_ah.shape[1],1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, GRU, TimeDistributed

model = Sequential()
model.add(GRU(32, input_shape = v_bahasa_ah.shape[1:], return_sequences = True))
model.add(TimeDistributed(Dense(8, activation = 'relu')))
model.add(TimeDistributed(Dense(total_kata_bahasa_uh+1, activation = 'softmax')))
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics =['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    v_bahasa_ah,
    v_bahasa_uh,
    batch_size = 2,
    epochs = 10,
    validation_split = 0.2
)
```

```python
test_kalimat = ['ah ahh ahhh uh uhh', 'ah ahh ih uh uhh']
test_kalimat = token_ah.texts_to_sequences(test_kalimat)
test_kalimat = pad_sequences(test_kalimat, maxlen = 20, padding ='post')
test_kalimat = test_kalimat.reshape(test_kalimat.shape[0], test_kalimat.shape[1], 1)
```

```python
def logits_to_text(logits, tokenizer):
  index_to_words = {id: word for word, id in tokenizer.word_index.items()}
  index_to_words[0] = ''
  return ' '.join([index_to_words[prediction] for prediction in np.argmax(logits, 1)])
```

```python
y_pred = model.predict(test_kalimat)
print(logits_to_text(y_pred[0], token_uh))
```
