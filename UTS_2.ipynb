```python
import pandas as pd
```

```python
test = pd.read_csv('UTS_test_soal_remedi.csv')
test
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(80, input_dim=15, activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(loss='mse', optimizer='adam',metrics=['mae'])
model.summary()
```

```python
model.load_weights('uts.h5')
```

```python
y_pred = model.predict(test)
```

```python
y_pred = y_pred.tolist()
y_pred = [y_pred[i][0] for i in range(len(y_pred))]
y_pred
```

```python
hasil_akhir = pd.DataFrame({"y" : y_pred})
hasil_akhir.to_csv("20102007.csv", index = False)
```
