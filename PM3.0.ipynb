```python
panjang = ahuh.shape[0]
batas = panjang*60//100
x = ahuh[ :, :-1]
y = ahuh[ : ,-1]
x_train = x[ :batas]
x_test = x[batas: ]
y_train = y[ :batas]
y_test = y[batas: ]
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
otak = LinearRegression()
```

```python
otak.fit(x_train, y_train)
```

```python
otak.predict([[2,1],[3,1],[1,2]])
```

```python
y_pred = otak.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

mae = mean_absolute_error(y_pred,y_test)
mse = mean_squared_error(y_pred,y_test,squared =False)
print(mae)
print(mse)
```
