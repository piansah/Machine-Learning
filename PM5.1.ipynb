```python
import glob
x = []
y = []
for name_file in glob.glob('image/air/*.jpg'):
    gambar = cv2.imread(name_file)
    gambar = cv2.resize(gambar,(100, 100))/255
    x.append(gambar)
    y.append(0)
```

```python
for name_file in glob.glob('image/tanah/*.jpg'):
    gambar = cv2.imread(name_file)
    gambar = cv2.resize(gambar,(100, 100))/255
    x.append(gambar)
    y.append(1)
```

```python
import numpy as np
import random
z = list(zip(x, y))
random.shuffle(z)
x = np.array(x)
y = np.array(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPool1D, MaxPooling2D
model = Sequential()
model.add(Conv2D(8, (3,3), input_shape =(100,100,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(16, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(
    x,
    y,
    batch_size = 2,
    epochs = 3,
    validation_split = 0.1
)
```

```python
x_test = []
gambar = cv2.imread('image/air/T_Air.jpg')
gambar = cv2.resize(gambar,(100,100))/255
x_test.append(gambar)
x_test = np.asarray(x_test)
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)*1
y_pred
```

```python
x_test = []
gambar = cv2.imread('image/tanah/T_Tanah.jpg')
gambar = cv2.resize(gambar,(100,100))/255
x_test.append(gambar)
x_test = np.asarray(x_test)
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)*1
y_pred
```
